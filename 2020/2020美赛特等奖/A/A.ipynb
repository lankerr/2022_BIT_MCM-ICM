## 2001334 
### 1. 海水模型是通过对1960-2019年的全球海洋温度数据进行分析，使用ARIMA模型，其中使用Bootstrap生成数据，模式识别这里需要注意的是，我们的话，在这里使用的
$$\Delta ^{(d)}u_{i,t} = \sum _{j=1}^p\Delta^{(d)}u_{i,t-j}+\epsilon_{i,t}$$
$$\hat{u}_{i,t}=E(\Delta u_{i,t})+\hat{u}_{i,t-1}+\epsilon_{i,t}$$
这里注意折交叉验证和我们的Bootstrap  https://zhuanlan.zhihu.com/p/24851814  这个用于在样本比较小的时候生成更多的样本
### 2. 还需要建立鱼类洄游和温度之间的关系，鱼类洄游的公式，这个
$$(dx,dy)=vdt$$
$$(x_t,y_t)=(x_{t-1},y_{t-1})+(dx,dy)$$
$$v(x,y,t)=f(\nabla u(x,y,t))$$
最后，所有的温度都得到了计算，然后是对(9)的沿严格近似下
$$v(x,y,t)=sign(\Delta u)\frac{\nabla u(x,y,t)}{||\nabla u||+\beta}v_{max}$$
然后采用编程，这里使用的是递归算法
